```python
import pandas as pd 
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
import datetime as dt
```

```python
df = pd.read_csv("/kaggle/input/corona-virus-report/country_wise_latest.csv")
```

```python
df.head()
```

```python
#keeping only required columns
df = df[['Country/Region', 'Confirmed', 'Deaths', 'Recovered', 'Active']]

#re-naming columns names
df.columns = ['country', 'confirmed', 'deaths', 'cured', 'active']
```

```python
df.head()
```

```python
#Sorting data w.r.t number of Deaths cases
max_deaths_cases = df.sort_values(by="deaths", ascending=False)
max_deaths_cases
```

```python
#Getting country with max number of deaths cases
top_country_deaths = max_deaths_cases[0:5]
```

```python
top_country_deaths
```

# Making bar-plot for country with top death cases


```python
sns.set(rc={'figure.figsize':(15,10)})
sns.barplot(x="country",y="deaths",data=top_country_deaths,hue="country")
plt.show()
```

```python
#Sorting data w.r.t number of confirmed  cases
max_confirmed_cases = df.sort_values(by="confirmed",ascending=False)
max_confirmed_cases
```

```python
#Getting country with max number of confirmed cases
top_country_confirmed_cases = max_confirmed_cases[0:5]
```

```python
top_country_confirmed_cases
```

# Making bar-plot for country with top confirmed cases


```python
#making bar-plot for country with top conirmed cases
sns.set(rc={'figure.figsize':(15,10)})
sns.barplot(x="country",y="confirmed",data=top_country_confirmed_cases,hue="country")
plt.show()
```

```python
#Sorting data w.r.t number of cured cases
cured_cases = df.sort_values(by="cured",ascending=False)
cured_cases
```

```python
#Getting country with max number of cured cases
cured_cases = cured_cases[0:5]
```

```python
cured_cases
```

# Making barplot for country and cured cases


```python
sns.set(rc={"figure.figsize":(15,10)})
sns.barplot(x="country",y="cured",data=cured_cases,hue="country")
plt.show()
```

```python
#sorting data w.r.t number of active cases
active_cases=df.sort_values(by="active",ascending=False)
active_cases
```

```python
active_cases=active_cases[0:5]
```

# Making barplot for country and active cases

```python
sns.set(rc={"figure.figsize":(15,10)})
sns.barplot(x="country",y="active",data=active_cases,hue="country")
plt.show()
```
